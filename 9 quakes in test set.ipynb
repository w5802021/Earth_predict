# spectral_contrast
The ```librosa``` library has the very nice feature function called ```spectral_contrast```, which we'll use to detect the segments with the quake peak inside. 

````python
%%time
PATH = "../input/LANL-Earthquake-Prediction/"
train = pd.read_pickle("../input/lanl-training-as-pickle/train.pkl")
````

````python
def plot_chunk(data, ax):
    ax.plot(data.acoustic_data, "C1", label=f"max aco_sig: {data.acoustic_data.max()}")
#     ax.set_xlabel('Index [-]')
    ax.set_ylabel('Acoustic signal', color='b')
    ax.tick_params('y', colors='b')
    ax.legend(loc=2)

    ax2 = ax.twinx()
    plt.plot(data.time_to_failure, "r")
#     plt.plot(data.acoustic_data.rolling(100).mean(), "C2", label="acoustic_rolling_mean")
    ax2.set_ylabel('Time to failure [s]', color='r')
    ax2.tick_params('y', colors='r')
#     ax2.set_ylim(0, 20)
    plt.legend()
    plt.title(f"chunksize = {data.shape[0]}")
````

## Train dataset

````python
length = 629145480
chunk = 150000
frames = int(length/chunk)

# idx = train.time_to_failure.values[:length:chunk][1:]#pd.DataFrame(train.time_to_failure.values[:length:chunk][1:], index=range(length), dtype=np.float64, columns=['time_to_failure'])

df = pd.DataFrame(index=range(frames))

for start in tqdm_notebook(range(frames)):
    data_chunk = train.loc[start*chunk:start*chunk+chunk]
    
    contrast = librosa.feature.spectral_contrast(data_chunk.acoustic_data.astype("float").values, hop_length=1501, fmin=200.0, n_bands=6, quantile=0.005, linear=True)
    df.loc[start, "contrast"] = contrast[0].max()
    df.loc[start, "time"] = start*chunk


fig, axes = plt.subplots(2,1, figsize=(25,15))
# 1% of signal
plot_chunk(train.loc[0:start*chunk+chunk:100], axes[0])

# # spectral_contrast
axes[1].plot(df.time, df.contrast.values, label="spectral_contrast")
axes[1].axhline(y=10000, color='k')
axes[1].text(0, 11000, "threshold = 10000", fontsize=15)
axes[1].legend()
plt.show();
````

## Test dataset

````python
submission = pd.read_csv("../input/LANL-Earthquake-Prediction/sample_submission.csv")

seg_gen = (pd.read_csv('../input/LANL-Earthquake-Prediction/test/' + seg_id + '.csv') for seg_id in submission.seg_id)
df = pd.DataFrame(index=range(2624))

for n,data_chunk in tqdm_notebook(enumerate(seg_gen)):
    
    contrast = librosa.feature.spectral_contrast(data_chunk.acoustic_data.astype("float").values, hop_length=1501, fmin=200.0, n_bands=6, quantile=0.005, linear=True)
    df.loc[n, "contrast"] = contrast[0].max()
    
print(f"{np.sum(df.contrast>10000)} peaks over the 10000 threshold.")

fig, axes = plt.subplots(1,1, figsize=(25,10))
# 1% of signal
# plot_chunk(train.loc[0:start*chunk+chunk:100], axes[0])

# # spectral_contrast
axes.plot(df.contrast.values, label="spectral_contrast")
axes.axhline(y=10000, color='k')
axes.text(0, 11000, "threshold = 10000", fontsize=15)
axes.legend()
plt.show();
````
